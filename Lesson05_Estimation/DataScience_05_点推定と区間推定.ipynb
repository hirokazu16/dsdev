## ライブラリをインポート

```python
# 必要なライブラリをimport
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

## 乱数を発生させる

```python
# 乱数種を設定
np.random.seed(1)
```

```python
# 平均15・分散1の正規分布に従う乱数を10000個発生させる
norm_dist = stats.norm(loc=15, scale=1)
data1 = norm_dist.rvs(size=10000)
```

```python
data1
```

## matplotlibでヒストグラムを作成

```python
# matplotlibでヒストグラムを作成
plt.hist(data1)
```

## seabornでヒストグラムと分布を作成

```python
# seabornでヒストグラムと分布を作成
sns.histplot(data1, kde=True)
```

## 母平均の点推定

```python
# 母平均の点推定（標本平均を母平均の推定値として扱う）
mu = np.mean(data1)
mu
```

##  母分散の点推定（不偏分散）

```python
# 母分散の点推定（不偏分散を母分散の推定値として扱う）
sigma2 = np.var(data1, ddof=1)
sigma2
```

## 標準誤差（SE）を計算

```python
# 標準誤差（SE）
sigma = np.std(data1, ddof=1)
se = sigma/np.sqrt(len(data1))
se
```

## 95%信頼区間を計算

```python
# 自由度（標本数 - 1）を計算
df = len(data1) - 1

# 95%信頼区間の下限値と上限値を計算
interval = stats.t.interval(0.95, df, loc=mu, scale=se)
interval
```
