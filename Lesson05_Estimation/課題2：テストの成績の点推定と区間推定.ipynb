```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

df = pd.read_csv("students_scores.csv")

df.head()
```

```python
english = df["english"]
```

```python
plt.hist(english)
```

```python
sns.histplot(english, kde=True)
```

```python
# 母平均
mu = np.mean(english)
mu
```

なぜ ddof=1にする必要があるのか

```python
# 母分散
sigma2 = np.var(english, ddof=1)
sigma2
```

```python
# 標準誤差
sigma = np.std(english, ddof=1)
se = sigma/np.sqrt(len(english))
se
```

```python
# 自由度（標本数 - 1）を計算
df = len(english) - 1

# 95%信頼区間の下限値と上限値を計算
interval = stats.t.interval(0.95, df, loc=mu, scale=se)
interval
```
