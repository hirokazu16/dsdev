```python
# コードセルでSQLを扱えるようにする
import sql
%load_ext sql
```

```python
# （インメモリ）データベースを起動して接続
%sql duckdb://
```

```python
# 全件表示できるようにする
%config SqlMagic.displaylimit = None
```

```python
%%sql
# penguinsテーブルをCSVファイルのデータから作成
CREATE TABLE penguins AS SELECT * FROM read_csv_auto('penguins.csv');
```

```python
%%sql
# penguinsテーブルを全件表示
SELECT * FROM penguins;
```

```python
%%sql
# species,body_mass_g,sexのカラムのみ取得
SELECT species, body_mass_g, sex FROM penguins;
```

```python
%%sql
# bill_length_mmが42.0のデータのみ取得
SELECT * FROM penguins WHERE bill_length_mm = 42.0;
```

```python
%%sql
# bill_length_mmが55.0以上のデータのみ取得
SELECT * FROM penguins WHERE bill_length_mm >= 55.0;
```

```python
%%sql
# bill_length_mmが55.0以上、かつ、body_mass_gが5000以上のデータのみ取得
SELECT * FROM penguins WHERE bill_length_mm >= 55.0 AND body_mass_g >= 5000;
```

```python
%%sql
# bill_length_mmが55.0以上、または、body_mass_gが5000以上のデータのみ取得
SELECT * FROM penguins WHERE bill_length_mm >= 55.0 OR body_mass_g >= 5000;
```

```python
%%sql
# bill_length_mmのデータが無い（NULL）もののみ取得
SELECT * FROM penguins WHERE bill_length_mm IS NULL;
```

```python
%%sql
# speciesが"Adelie"、かつ、islandが"Biscoe"のデータのみ取得
SELECT * FROM penguins WHERE species = 'Adelie' AND island = 'Biscoe';
```

```python
%%sql
# body_mass_gの昇順で並べ替え
SELECT * FROM penguins WHERE species = 'Adelie' AND island = 'Biscoe' ORDER BY body_mass_g;
```

```python
%%sql
# body_mass_gの降順で並べ替え
SELECT * FROM penguins WHERE species = 'Adelie' AND island = 'Biscoe' ORDER BY body_mass_g DESC;
```

```python
%%sql
# body_mass_gの降順を第一に、同値の場合はbill_length_mmの降順で並べ替え
SELECT * FROM penguins WHERE species = 'Adelie' AND island = 'Biscoe' ORDER BY body_mass_g DESC, bill_length_mm DESC;
```

```python
%%sql
# 全件の件数を取得
SELECT COUNT(*) FROM penguins;
```

```python
%%sql
# 全件の件数の列名に"number_of_penguins"という別名をつけて表示
SELECT COUNT(*) AS number_of_penguins FROM penguins;
```

```python
%%sql
# bill_length_mmが42.0のデータの件数を取得
SELECT COUNT(*) FROM penguins WHERE bill_length_mm = 42.0;
```

```python
%%sql
# body_mass_gの合計を取得
SELECT SUM(body_mass_g) FROM penguins;
```

```python
%%sql
# body_mass_gの平均を取得
SELECT AVG(body_mass_g) FROM penguins;
```

```python
%%sql
# speciesごとの件数を取得
SELECT species, COUNT(*) FROM penguins GROUP BY species;
```

```python
%%sql
# speciesごとの件数を取得し、その数が150以上のもののみ表示
SELECT species, COUNT(*) AS number_of_penguins FROM penguins GROUP BY species HAVING number_of_penguins >= 150;
```

```python
# bill_length_mmが42.0のデータのみを抽出してresult.csvへ保存
result = %sql SELECT * FROM penguins WHERE bill_length_mm = 42.0
result.csv(filename="result.csv")
```
